```python

driver = webdriver.Chrome("chromedriver.exe")
for leagueName, archiveName, layoutType, leagueId in leagues:
    
    print("opening the browser...")
    
    
    url = 'https://www.google.com/search?q=campeonato+brasileiro+final+2018+google+resultados#sie=lg;'+leagueId+';2;/m/0fnk7q;mt;fp;1;;'
    
    # driver utilizado para interacao com o Google Chrome
    
    driver.get(url) 
    driver.set_page_load_timeout(10)
    driver.maximize_window()
    driver.implicitly_wait(10)
    print("Entering the league page...")
    #scrolls page to top, the page has a top scroll with ajax loading content

    body = driver.find_element_by_css_selector('body')
    body.click()
    print("Scrolling up...")
    for i in range(20):
        print("Scrolling "+str(i)+" of 20 times")
        time.sleep(2)
        body.send_keys(Keys.PAGE_UP) 
        body.send_keys(Keys.PAGE_UP)
        body.send_keys(Keys.PAGE_UP)
        body.send_keys(Keys.PAGE_UP)
    #data-mid esse Ã© o id das partidas para alterar no link

    elements = driver.find_elements_by_xpath('//*[@data-mid]')

    #get match link ids
    match_ids = []
    for match in elements:
        #print(match.find_element_by_xpath('..').find_element_by_xpath('..').find_element_by_xpath('..').find_element_by_xpath('..').find_element_by_xpath('..').find_element_by_xpath('..').find_element_by_xpath('..').find_element_by_xpath('..').find_element_by_xpath('..').find_element_by_xpath('..').find_element_by_xpath('//div/0').text)
        #print(match.get_attribute("data-mid"))
        match_ids.append(match.get_attribute("data-mid"))
    
    data =[]
    err = []
    print("Starting scraping")
    for match in match_ids:
        try:
            data.append(getDataFromGoogleGamePage(driver, match, leagueName, layoutType))
        except:
            try:
                data.append(getDataFromGoogleGamePage(driver, match, leagueName, 1))
            except:
                print("erro matchid="+match)
                err.append(match)

    df_ = pd.DataFrame(data, columns = columns,)

    df_.head()
    
    path = 'data/'+archiveName
    df_.to_csv(path, index = None, header=True,encoding="utf-8")
```
